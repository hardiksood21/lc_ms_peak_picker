## Download Test Data
We'll use a public metabolomics dataset.

```r
# Download a small test mzML file from xcms test data
download.file("https://raw.githubusercontent.com/sneumann/xcms/master/tests/mzML/MM14.mzML", 
              "test_data.mzML")
cat("✓ Test data downloaded\n")
```

## Run Peak Detection

Using the same CentWave parameters as the Galaxy tool:

```r
# Read data
raw_data <- readMSData("test_data.mzML", mode = "onDisk")

# Set CentWave parameters (matching Galaxy tool defaults)
cwp <- CentWaveParam(
    ppm = 15,
    peakwidth = c(5, 20),
    snthresh = 1.5,
    prefilter = c(0, 0)
)

# Detect peaks
xdata <- findChromPeaks(raw_data, param = cwp)
peaks <- as.data.frame(chromPeaks(xdata))

cat("✓ Detected", nrow(peaks), "peaks\n")
head(peaks)
```
